```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import holoviews as hv
 
from bokeh.io import output_notebook
output_notebook()

```

```python
# Load csv file into datafram
ca_county_df = pd.read_csv("Resources/ca_county_data.csv")

# ca_county_df.dtypes

# Display df
ca_county_df
```

```python
%%capture --no-display

# Configure map
ca_map_plot = ca_county_df.hvplot.points(
    "Lng", 
    "Lat",
    geo=True,
    color="District",
    tiles = "EsriNatGeo",
    size="Beef cows (head)",
    hover_cols=["District", "County", "County Seat", "Beef cows (% of total)"],
    frame_width=600,
    frame_height=600,
    scale = 0.15,
    title="California County Map",
    legend="top_right"
    )


# Display map
ca_map_plot
```

Unsuprisingly, San Joaquin Valley and Sacramento Valley have the largest number of cattle. The weather is more consistent and the valleys provide food through pastures and grasses etc. that grow in the open areas.
